**Enigma**                                                                       
the main parts of enigma are:
*   ***Interface***, with a keyboard, when a button is pressed a lamp lights the encrypted letter, so it represent the input and the output in the same time, in this code we will divide into input and output the interface
*   ***Stackerbrett***, is a tableau of sockets, a mechanism created to make the decryption harder, it simply map letters into others, it takes directly the input, is the first layer
*   ***Rotors***, these are mechanism that physically implement the caesar cipher with some complications: let's assume that we have 3 rotors and let's represent the K (shift of the caesar cipher) as a tuple (K1,K2,K3). ie if we have (0,0,0) the letter A will be encrypted as A because there is no shift of the rotors. if we have (1,0,0) A->B and so on. Now the complications. every time the first rotor does 26 rotation, the second does 1 rotation, in the same way at the 26th rotation of the second rotor the third does a rotation. But the second does a turn also when the third does one. And the rotor alpha rotate every time the stackerbrett is activated
*    ***Reflector***, reverse the text

the encryption path is:

interface -> stackerbrett -> rotors (1 -> 2 -> 3) -> reflector -> rotors (3 -> 2 -> 1)
